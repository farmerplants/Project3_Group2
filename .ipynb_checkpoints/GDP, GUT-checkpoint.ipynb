```python
import numpy as np
import pandas as pd
from pathlib import Path
from functools import reduce
from matplotlib import pyplot as plt
import hvplot.pandas
```

```python
col_nums_value=[]
for x in range(3,14):
    col_nums_value.append(x)
```

```python
# Read in ZHVI
file_path = Path("data/rateofgrowth.csv")
GTM1 = pd.read_csv(file_path, parse_dates=col_nums_value)
GTM1.head()
```

```python
# traspose the data
df = pd.melt(GTM1, id_vars = ['Series Name','Country Name', "Country Code" ], var_name='Years', value_name='GDPGrowth')
```

```python
col_nums_value1=[]
for x in range(4,66):
    col_nums_value.append(x)
```

```python
# Read in ZHVI
file_path = Path("data/GDP1.csv")
GTM = pd.read_csv(file_path, parse_dates=col_nums_value1)
GTM.head()
```

```python
# traspose the data
df1 = pd.melt(GTM, id_vars = ['Country Name', "Country Code", "Indicator Name", "Indicator Code"], var_name='Years',value_name='GDP' )
```

```python
df1
```

```python
df1.drop(columns=['Indicator Code'])
```

```python
GMT1 = df1.loc[(df1['Country Name']=='Guatemala'), ['Country Name', 'Country Code', 'Indicator Name','Years', 'GDP']]
GMT1.dropna(inplace=True)
GMT1["GDP"]=round(((GMT1["GDP"]/1000000000)),2)
GMT1
```

```python
GMT1.plot.bar(x ='Years', y = 'GDP', title = "Guatemala GDP from 1960 to 2020",figsize=(15,7.5))
```
